# Exerc√≠cio 9 - C√°lculo de Tokens em Texto Longo com API Gemini

**Enunciado:** Implemente um **notebook** que use a API Gemini para calcular a quantidade de **tokens** necess√°rios para processar um texto de 5.000 palavras. Baseie-se no modelo de tokeniza√ß√£o utilizado por Gemini e explique como a quantidade de tokens influencia o custo e o desempenho da intera√ß√£o com LLMs em textos longos.

## Baixando um arquivo pdf longo

````python
import requests
import fitz  # PyMuPDF

# URL do PDF
url = "https://estudosaudiovisuais.wordpress.com/wp-content/uploads/2013/05/cidadededeus12.pdf"
# Nome do arquivo para salvar o PDF baixado
pdf_path = "cidadededeus12.pdf"

# Baixar o PDF
response = requests.get(url)
with open(pdf_path, "wb") as file:
    file.write(response.content)

print("PDF baixado com sucesso!")

# Fun√ß√£o para extrair o texto do PDF
def extract_text_from_pdf(pdf_path):
    with fitz.open(pdf_path) as doc:
        text = ""
        for page_num in range(doc.page_count):
            page = doc[page_num]
            text += page.get_text()
    return text

# Extrair o texto
pdf_text = extract_text_from_pdf(pdf_path)

texto_5000_palavras = " ".join(pdf_text.split()[:5000])
````

## Calculando o n√∫mero de Tokens

````python
# Exerc√≠cio 9: C√°lculo de Tokens
import math

def calcular_tokens_gemini(texto):
    num_tokens = 0
    max_palavras = 1000
    tamanho_texto = len(texto.split())
    num_partes = math.ceil(tamanho_texto/max_palavras)
    for i in range(num_partes):
        print(f"Parte: {i}")
        idx_inicio = i * max_palavras
        idx_fim = max_palavras * (i + 1)
        texto_parte = " ".join(texto.split()[idx_inicio:idx_fim])
        # print(f"Parte: {texto_parte}")
        embeddings = genai.embed_content(
            model="models/text-embedding-004",
            content=texto_parte,
            task_type="retrieval_document",
            title="Embedding of single string"
        )
        num_tokens += len(embeddings['embedding'])

    return num_tokens

print(f"N√∫mero de palavras no texto: {len(texto_5000_palavras.split())}\n")

tokens = calcular_tokens_gemini(texto_5000_palavras)

print("\nTotal de tokens estimados:", tokens)
````

A quantidade de tokens influencia diretamente o custo e o desempenho da intera√ß√£o com modelos de linguagem (LLMs) porque cada token representa uma unidade m√≠nima de texto (palavras ou fragmentos de palavras) processada pelo modelo. Modelos como o Gemini ou o GPT cobram de acordo com o n√∫mero de tokens enviados e recebidos, o que significa que quanto mais longo o texto, maior ser√° o custo. Otimizar o n√∫mero de tokens enviados ao modelo n√£o s√≥ reduz custos, mas tamb√©m melhora a velocidade de resposta, tornando a intera√ß√£o mais eficiente para o usu√°rio.

# Exerc√≠cio 12 - Sistema de Envio de Emails com Dados Tur√≠sticos em JSON

**Enunciado:** Queremos montar um sistema que envie emails com os pontos tur√≠sticos de cidades num roteiro de viagem. Com base no **Exerc√≠cio 11**, altere o prompt para estruturar a sa√≠da em formato JSON, com descri√ß√µes e n√∫mero de visitantes. Crie um **notebook** que use a biblioteca pandas para converter esse JSON para dataframe e plot um √∫nico gr√°fico de barras horizontais com todos os pontos tur√≠sticos, destacando as cores das barras de acordo com a cidade do ponto tur√≠stico (Coliseu, em Roma deve ter uma cor diferente que Parthenon, em Atenas). Coloque as legendas no gr√°fico.

````python
# Imports necess√°rios
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

def extrair_entre_colchetes(texto):
    # Usa express√£o regular para encontrar o conte√∫do incluindo os colchetes
    resultado = re.search(r'\[.*\]', texto, re.DOTALL)
    if resultado:
        return resultado.group(0)  # Retorna a correspond√™ncia inteira, incluindo os colchetes
    return None


# Estrutura do prompt com YAML para a estrutura de sa√≠da em JSON
prompt_yaml = """
roteiro:
  roma:
    descricao: "Liste 3 pontos tur√≠sticos de Roma com descri√ß√µes e n√∫mero anual de visitantes. Sa√≠da no formato JSON, incluindo 'ponto_turistico', 'descricao' e 'visitantes_anuais'. 'visitantes_anuais' tem que estar no formato n√∫merico, i.e. 1000000 se for 1 milh√£o"
  atenas:
    descricao: "Liste 3 pontos tur√≠sticos de Atenas com descri√ß√µes e n√∫mero anual de visitantes. Sa√≠da no formato JSON, incluindo 'ponto_turistico', 'descricao' e 'visitantes_anuais'. 'visitantes_anuais' tem que estar no formato n√∫merico, i.e. 1000000 se for 1 milh√£o"
"""

# Carrega o YAML em um dicion√°rio Python
roteiro = yaml.safe_load(prompt_yaml)

# Armazena os roteiros em uma lista para processar no gr√°fico
roteiros = []

# Gera o roteiro para cada cidade
for local, detalhes in roteiro['roteiro'].items():
    print(f"Roteiro para {local.capitalize()}:")
    roteiro_local = chamar_api_gemini(detalhes['descricao'])
    roteiro_formatado = json.loads(extrair_entre_colchetes(roteiro_local))
    roteiros.extend([{"cidade": local.capitalize(), **ponto} for ponto in roteiro_formatado])

# Cria um DataFrame a partir do JSON de todos os pontos tur√≠sticos
df = pd.DataFrame(roteiros)

# Configura o estilo do gr√°fico
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.color_palette("Set2")

# Constr√≥i o gr√°fico de barras horizontais
plot = sns.barplot(
    data=df,
    y='ponto_turistico',
    x='visitantes_anuais',
    hue='cidade',
    dodge=False
)

# Configura t√≠tulos e legendas
plt.title('Pontos Tur√≠sticos por N√∫mero Anual de Visitantes')
plt.xlabel('Visitantes Anuais')
plt.ylabel('Ponto Tur√≠stico')
plt.legend(title='Cidade')
plt.show()
````
